```python
import matplotlib.pyplot as plt

# Datos
x = [1000, 10000, 100000, 1000000]

# Tiempos para cada algoritmo
selection_sort_y = [719, 70766, 7122386, 717148713]
merge_sort_y = [154, 1823, 18895, 200158]
quick_sort_y = [17, 242, 3066, 37136]
cpp_sort_y = [37, 525, 5550, 70254]

# Crear la gráfica
plt.figure(figsize=(10, 5))

# Graficar cada algoritmo
plt.plot(x, selection_sort_y, marker='o', color='m', linestyle='-', label="Selection Sort")
plt.plot(x, merge_sort_y, marker='o', color='b', linestyle='-', label="Merge Sort")
plt.plot(x, quick_sort_y, marker='o', color='r', linestyle='-', label="Quick Sort")
plt.plot(x, cpp_sort_y, marker='o', color='y', linestyle='-', label="Sort C++")

# Escala logarítmica
plt.xscale('log')
plt.yscale('log')

# Etiquetas y título
plt.xlabel('Tamaño de datos')
plt.ylabel('Tiempo en microsegundos')
plt.title('Comparación de Algoritmos de Ordenamiento')

# Cuadrícula y leyenda
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.show()
```
